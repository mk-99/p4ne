```python
import requests, json, pprint, ssl
from urllib3.poolmanager import PoolManager
from requests.adapters import HTTPAdapter

class Ssl1HttpAdapter(HTTPAdapter):
    def init_poolmanager(self, connections, maxsize, block=False):
        self.poolmanager = PoolManager(num_pools=connections, maxsize=maxsize, block=block,\
                                       ssl_version=ssl.PROTOCOL_TLSv1)

requests.packages.urllib3.util.ssl_.DEFAULT_CIPHERS = 'ALL:@SECLEVEL=1'
net_session = requests.Session()
net_device = "https://10.31.70.210:55443"
net_session.mount(net_device, Ssl1HttpAdapter())
```

```python
# Credentials for test network device
login = 'restapi'
password = 'j0sg1280-7@'
r = net_session.post(net_device + '/api/v1/auth/token-services', auth=(login, password), verify=False)
print(r.status_code)
```

```python
r.json()
token = r.json()['token-id']
token
```

```python
headers = {
    "content-type": "application/json",
    "X-Auth-Token": token
}
```

```python
url = net_device + "/api/v1/global/memory/processes"
r = net_session.get(url, headers=headers, verify=False)
```

```python
processes = r.json()['processes']
ps_sorted = sorted(processes, key=lambda ps: int(ps['memory-used']), reverse=True)
ps_sorted[0:10]
```
